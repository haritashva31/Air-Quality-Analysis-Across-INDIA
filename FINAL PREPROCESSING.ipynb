```python
import os
import pandas as pd

input_folder_path = "/Users/meetsmacbook/Downloads/chaljaa"
output_folder_path = "/Users/meetsmacbook/Downloads/cmon"
csv_files = [file for file in os.listdir(input_folder_path) if file.endswith(".csv")]
common_columns = ['CO (mg/m3)', 'To Date', 'NO (ug/m3)', 'PM10 (ug/m3)', 'Ozone (ug/m3)', 
                  'Benzene (ug/m3)', 'NO2 (ug/m3)', 'PM2.5 (ug/m3)', 'NOx (ppb)', 
                  'From Date', 'SO2 (ug/m3)']
file_years = {}
for file in csv_files:
    input_file_path = os.path.join(input_folder_path, file)
    df = pd.read_csv(input_file_path)
    df_common = df[common_columns]
    df_common = df_common.fillna(df_common.mean(numeric_only=True))
    df_common = df_common.replace(0, df_common.mean(numeric_only=True))
    df_common = df_common.apply(lambda x: x.fillna(x.mean(numeric_only=True)) if x.dtype.kind in 'biufc' else x)

    from_date = pd.to_datetime(df_common['From Date']).dt.year.min()
    to_date = pd.to_datetime(df_common['To Date']).dt.year.max()
    file_years[file] = (from_date, to_date)
    output_file_path = os.path.join(output_folder_path, f"preprocessed_{file}")
    df_common.to_csv(output_file_path, index=False)

    print(f"Processed and Preprocessed: {file}, Saved to: {output_file_path}")
    print(f"Data Range: {from_date} to {to_date}")

print("\nRange of Years for Each File:")
for file, years in file_years.items():
    print(f"{file}: {years[0]} to {years[1]}")
```

```python
import os
import pandas as pd
input_folder_path = "/path/to/your/input/folder"
output_folder_path = "/path/to/your/output/folder"
csv_files = [file for file in os.listdir(input_folder_path) if file.endswith(".csv")]
common_columns = ['From Date', 'To Date','CO (mg/m3)','NO (ug/m3)', 'PM10 (ug/m3)', 'Ozone (ug/m3)', 
                  'Benzene (ug/m3)', 'NO2 (ug/m3)', 'PM2.5 (ug/m3)', 'NOx (ppb)','SO2 (ug/m3)']
file_years = {}

for file in csv_files:
    input_file_path = os.path.join(input_folder_path, file)
    df = pd.read_csv(input_file_path)
    df = df.fillna(df.mean(numeric_only=True))
    df = df.replace(0, df.mean(numeric_only=True))
    df = df.apply(lambda x: x.fillna(x.mean(numeric_only=True)) if x.dtype.kind in 'biufc' else x)
    from_date = pd.to_datetime(df['From Date']).dt.year.min()
    to_date = pd.to_datetime(df['To Date']).dt.year.max()
    file_years[file] = (from_date, to_date)
    ordered_columns = ['From Date', 'To Date'] + [col for col in common_columns if col not in ['From Date', 'To Date']]
    df_ordered = df[ordered_columns]
    output_file_path = os.path.join(output_folder_path, file)
    df_ordered.to_csv(output_file_path, index=False)
    print(f"Processed and Preprocessed: {file}, Saved to: {output_file_path}")
    print(f"Data Range: {from_date} to {to_date}")

print("\nRange of Years for Each File:")
for file, years in file_years.items():
    print(f"{file}: {years[0]} to {years[1]}")
```
